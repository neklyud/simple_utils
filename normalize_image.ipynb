```python
import torch
import torchvision.transforms as transforms
import torchvision.datasets as datasets
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np
import random
from PIL import Image
```

```python
%matplotlib inline
def show(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1,2,0)), interpolation='nearest')
```

```python
import scipy.misc
from scipy.ndimage.filters import gaussian_filter
from scipy.ndimage.filters import uniform_filter
import skimage
from numpy.random import normal 
```

```python
blur = lambda img: gaussian_filter(img, sigma=(0.9, 0.9, 0))
noise = lambda img: torch.add(img, torch.Tensor(img.data.new(img.size()).normal_(0, 0.1)))
```

```python
lena = scipy.misc.face()
print(lena.shape)
np.random.seed(0)
im = Image.fromarray(np.uint8(lena))
img = transforms.Resize((512, 512))(im)
img = transforms.RandomCrop((506))(img)
img = transforms.Resize((512, 512))(img)
img = transforms.ColorJitter(brightness=np.random.uniform(0,0.3))(img)
img = transforms.Lambda(blur)(img)
img = transforms.ToPILImage()(img)
img = transforms.ToTensor()(img)
#img = transforms.Lambda(noise)(img)

print(img.size())
```
